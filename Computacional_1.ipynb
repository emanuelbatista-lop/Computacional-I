###Valores do intervalo da integral

```python
x_i = 1
x_f = 5
delta_x = 0.5
intervalo_1_5 = []
for i in range(int((x_f-x_i)/delta_x)):
  intervalo_1_5.append(x_i+i*delta_x)

print(f'Apresentação dos 5 primeiros valores do intervalo: {intervalo_1_5[:5]}')
```

###Integral analítica e somatória

```python
x_i = 1
x_f = 5
delta_x = 0.001
intervalo_1_5 = []
for i in range(int((x_f-x_i)/delta_x)):
  intervalo_1_5.append(x_i+i*delta_x)

fxi = []
ci_fxi = []
for i in range(len(intervalo_1_5)):
  f_x_i = 4*intervalo_1_5[i]**2 + 6*intervalo_1_5[i] - 8
  ci_fxi.append(f_x_i*delta_x)
  fxi.append(f_x_i)
#f(x) = 4x² + 6x - 8
integral_analitico = (4/3)*(5**3-(1)**3) + 3*(5**2-(1)**2) - 8*(5-1)

print(f'Apresentação dos 5 primeiros valores do somatório: \n{ci_fxi[:5]}\n')
print(f'Resultado da integral calculada pela somatória: {sum(ci_fxi):.2f}')
print(f'Resultado da integral calculada analiticamente: {integral_analitico:.2f}')
```

####Gráfico da integral analítica

```python
plt.figure(figsize = (10,6))
plt.plot(intervalo_1_5, fxi, 'b', label = r'$f(x) = 4x^2 + 6x + 8$')

for i in range(len(intervalo_1_5)):
  x0 = intervalo_1_5[i]
  plt.bar(x0, fxi[i], width=delta_x, align='edge', color='orange', edgecolor='black', alpha=0.5)

plt.title('Integração numérica')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

###Integral do ponto médio


```python
x_i = 1
x_f = 5
delta_x = 0.1
intervalo_1_5 = []
for i in range(int((x_f-x_i)/delta_x)):
  intervalo_1_5.append(x_i+i*delta_x)

fx_medio = []
for i in range(len(intervalo_1_5)):
  x_medio = intervalo_1_5[i] + delta_x / 2
  f_x_medio = (4*x_medio**2) + (6*x_medio) - 8
  fx_medio.append(f_x_medio)

integral_ponto_medio = delta_x * sum(fx_medio)
integral_analitico = (4/3)*(5**3 - 1**3) + 3*(5**2 - 1**2) - 8*(5-1)

print(f'Resultado pela integral do Ponto Médio: {integral_ponto_medio:.2f}')
print(f'Resultado pela integral analítica: {integral_analitico:.2f}')
```

###Integral - Regra do trapézio

```python
x_i = 1
x_f = 5
delta_x = 0.05
intervalo_1_5 = []
for i in range(int((x_f-x_i)/delta_x) + 1):
  intervalo_1_5.append(x_i+i*delta_x)

fxi=[]
for i in range(len(intervalo_1_5)):
  x0 = intervalo_1_5[i]
  f_x = 4*(x0**2) + 6*(x0) - 8
  fxi.append(f_x)

f_a = fxi[0]
f_b = fxi[-1]
f_a_b = 2 * sum(fxi[1:-1])

integral_trapezio = (delta_x/ 2) * (f_a + f_a_b + f_b)

integral_analitico = (4/3)*(5**3 - 1**3) + 3*(5**2 - 1**2) - 8*(5-1)

print(f'Valores de f(x): {fxi}')
print(f'Resultado pela regra do trapézio: {integral_trapezio:.2f}')
print(f'Resultado analítico: {integral_analitico:.2f}')
```

####Gráfico da Regra do Trapézio

```python
plt.figure(figsize=(10,6))

intervalo = intervalo_1_5


# Construindo os trapézios manualmente
for i in range(len(intervalo) - 1):
    x0 = intervalo[i]
    x1 = intervalo[i+1]
    y0 = fxi[i]
    y1 = fxi[i+1]

    # Desenhando as quatro linhas do trapézio
    plt.plot([x0, x0], [0, y0], 'k', linewidth=1)      # linha vertical esquerda
    plt.plot([x1, x1], [0, y1], 'k', linewidth=1)      # linha vertical direita
    plt.plot([x0, x1], [y0, y1], 'k', linewidth=3)     # linha inclinada entre f(x0) e f(x1)
    plt.plot([x0, x1], [0, 0], 'k', linewidth=1)       # base no eixo x

intervalo_grafico = []

for i in range(int((x_f - x_i)/(delta_x/100))):
  intervalo_grafico.append(x_i + i*delta_x/100)


fxi_g = []
for i in range(len(intervalo_grafico)):
  f_x_i = 4*(intervalo_grafico[i])**2 + 6*(intervalo_grafico[i]) - 8
  fxi_g.append(f_x_i)
plt.plot(intervalo_grafico, fxi_g, 'y', label = 'f(x) = 4x² + 6x - 8')  # curva da função

plt.title('Integração Numérica - Regra do Trapézio')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

###Integral de Simpson

```python
x_i = 1
x_f = 5
delta_x = 0.5
n = int((x_f - x_i) / delta_x)

if n % 2 != 0:
    n += 1

x = [x_i + i * delta_x for i in range(n + 1)]

def f(x):
    return 4*x**2 + 6*x - 8

fx = [f(valor) for valor in x]

pares = fx[2:-1:2]
impares = fx[1:-1:2]

integral_simpson = (delta_x / 3) * (fx[0] + 4*sum(impares) + 2*sum(pares) + fx[-1])

integral_analitico = (4/3)*(x_f**3 - x_i**3) + 3*(x_f**2 - x_i**2) - 8*(x_f - x_i)

print(f"Integral (Simpson Composta): {integral_simpson:.6f}")
print(f"Integral analítica (exata): {integral_analitico:.6f}")
print(f"Erro absoluto: {abs(integral_simpson - integral_analitico):.8f}")
```

###Desse pros próximos dois, Regra da Extremidade Esquerda

```python
def Regra_1 (h = 0.5, x_i = 1, x_f = 5, a = 4, b = 6, c = -8):
  intervalo_1_5 = []

  for i in range(int((x_f - x_i)/h)):
    intervalo_1_5.append(x_i + i*h)

  fxi = []
  ci_fxi = []
  for i in range(len(intervalo_1_5)):
    x = intervalo_1_5[i]
    f_x_i = a*x**2 + b*x + c
    ci_fxi.append(f_x_i*h)
    fxi.append(f_x_i)

  integral_numerica = sum(ci_fxi)
  return integral_numerica
```

```python
from matplotlib.patches import BoxStyle
import matplotlib.pyplot as plt
def Regra_Extremidade_Esquerda (a = 0, b = 0, c = 0, d = 4, e = 6, f = -8, x_i = 1, x_f = 5, passo = 0.25, plot = True):
  intervalo = []
  for i in range(int((x_f - x_i)/passo)):
    intervalo.append(x_i + i*passo)

  fxi = []
  ci_fxi = []
  for i in range(len(intervalo)):
    x = intervalo [i]
    f_x_i = a*x**5 + b*x**4 + c*x**3 + d*x**2 + e*x + f
    ci_fxi.append(f_x_i*passo)
    fxi.append(f_x_i)

  somatoria = sum(ci_fxi)

  integral_xi = (a/6)*x_i**6 + (b/5)*x_i**5 + (c/4)*x_i**4 + (d/3)*x_i**3 + (e/2)*x_i**2 + f*x_i
  integral_xf = (a/6)*x_f**6 + (b/5)*x_f**5 + (c/4)*x_f**4 + (d/3)*x_f**3 + (e/2)*x_f**2 + f*x_f
  integral_analitico = integral_xf - integral_xi

  ostra = {
      'Intervalo': intervalo,
      'f(x)': fxi,
      'Integral Numérica': somatoria,
      'Integral Analítica': integral_analitico}

  if plot == True:
    plt.figure(figsize=(10,6))
    plt.plot(intervalo, fxi, 'b', label=rf'$f (x) = {a}x^5 + {b}x^4 + {c}x^3 + {d}x^2 + {e}x + {f}$')

  for i in range(len(intervalo)):
    x0 = intervalo[i]
    plt.bar(x0, fxi[i], width=passo, align='edge', color='orange', edgecolor='black', alpha=0.5)

  plt.text(0.3, 0.15, rf'$\int_{{{x_i}}}^{{{x_f}}} f(x)\, dx = \int_{{{x_i}}}^{{{x_f}}} {a}x^5 + {b}x^4 + {c}x^3 + {d}x^2 + {e}x + {f}\, dx = {somatoria:.2f}$',
           transform=plt.gca().transAxes,
           fontsize=12,
           verticalalignment='top',
           bbox=dict(boxstyle="round,pad=0.3", facecolor='white'))

  return ostra

Regra_Extremidade_Esquerda(plot=False)
```

###Aplicação da derivação de 1° ordem na forma numérica

#####Caso polinomial

######Esse é o calculo da segunda derivada: f′′(x) ≈ (Δx) -2f*(x+Δx) − 2f*(x) + f*(x−Δx) / (Δx)²​.

o x+delta_x = Xk+1 e vice versa

```python
delta_x = 5
x = 1

derivada = ((4*(x+delta_x)**2 + 6*(x+delta_x) - 8) - (4*(x-delta_x)**2 + 6*(x-delta_x) - 8))/(2*delta_x)
derivada_real = 8*x + 6
segunda_derivada = ((4*(x+delta_x)**2 + 6*(x+delta_x) - 8) - 2*(4*x**2 + 6*x - 8) + (4*(x-delta_x)**2 + 6*(x-delta_x) - 8))/(delta_x**2)
segunda_derivada_real = 8

print(f"Derivada: {derivada:.3f}")
print(f"Derivada Real: {derivada_real:.3f}")
print(f"Segunda Derivada: {segunda_derivada:.3f}")
print(f"Segunda Derivada Real: {segunda_derivada_real:.3f}")
```
