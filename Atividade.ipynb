##Atividade 1 - def Integral de Simpson

```python
import numpy as np

#Função original
def f(x):
  return (-3)*x**2 + 5*x - 7

#Integral analítica
def F(x):
  return (-9/3)*x**3 + 3*x**2 - 7*x

#Regra de Simpson Composta
def simpson(a, b, n):
  if n % 2 != 0:
    raise ValueError("n deve ser par para a Regra de Simpson.")

  h = (b - a)/n
  x = np.linspace(a, b, n+1)
  y = f(x)
  soma_2 = np.sum(y[2:-1:2])
  soma_4 = np.sum(y[1:-1:2])

  integral_numerica = (h/3)*(y[0] + y[-1] + 4*soma_4 + 2*soma_2)
  integral_analitica = F(b) - F(a)

  return integral_numerica, integral_analitica, x, y

a = 1
b = 5
n = 2

resultado, resultado_real, x, y = simpson(a, b, n)

print("Integral Númerica (Simpson):", resultado)
print("Integral Analítica:", resultado_real)
```

##Atividade 2 - def Derivada de uma função Algébrica


```python
#Função = 4x² + 6x − 8

def derivada1(f, x, h=1):
  return ((f(x + h) - f(x - h)) / (2*h))

def derivada2(f, x, h=1):
  return ((f(x + h) - 2*f(x) + f(x - h)) / (h**2))

def f(x):
  return 4*x**2 + 6*x - 8

print(derivada1(f, 2))
print(derivada2(f, 2))
```

##Atividade 3 - def Regressão Linear por Matrizes

```python
from numpy.linalg import norm
import numpy as np

def reg_lin_matriz(x,y):
  x = np.array(x).reshape(-1,1)
  y = np.array(y).reshape(-1,1)

  X = np.hstack([np.ones_like(x), x])
  beta = np.linalg.inv(X.T @ X) @ X.T @ y

  a = beta[0,0]
  b = beta[1,0]

  return a, b

x = [2, 4, 6]
y = [1, 3, 5]

a, b = reg_lin_matriz(x, y)
print("Intercepto (a): ", a)
print("Coeficiente angular (b):", b)
```
